```python
X= np.linspace(0,L,101)
Y= np.zeros(X.shape)
for i,v in enumerate(X):
    Y[i] = (v/2)-((v**3)/6)
plt.plot(X,Y,label="exact",linestyle="dashed")
plt.legend()

Intervals = np.array([5,10,20,50,100])

for n in Intervals:
    matrix = np.zeros((n-1,n-1))
    vector = np.zeros(n-1)
    
    x= np.linspace(0,L,n+1)[1:-1]
    
    for i in range(0,n-1):
        vector[i]= (1/n)*(i+1)
        
        for j in range(0,n-1):
            if i==j:
                matrix[i][j]= 2/((1/n)**2)
            elif i+1== j:
                matrix[i][j]= -1/((1/n)**2)
            elif i-1 ==j:
                matrix[i][j]= -1/((1/n)**2)
            else:
                matrix[i][j]= 0
        if i== n-2:
            matrix[i][n-2]= 1/((1/n)**2)
            
    y= np.linalg.solve(matrix,vector)
    
    plt.plot(x,y,label="Intervals="+ str(n))
    plt.legend()

plt.show()
  
```
